```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Spectrometer-1 Readings below

Below, we have two columns, the first column is just the index

```python
trtfile = pd.read_csv(
    "GROUPI_LAB0001.trt"
)

display(trtfile)

data = pd.read_csv(
    "GROUPI_LAB0001.trt",
    sep=";",
    skiprows=8,          # skip header lines
    names=["Wavelength_nm", "Counts"]
)

plt.figure()
plt.plot(data["Wavelength_nm"], data["Counts"])
plt.xlabel("Wavelength (nm)")
plt.ylabel("Counts")
plt.title("M-Normal Source On")
plt.savefig("M-NormalSource_On.jpg", dpi=400)
plt.show()
```
