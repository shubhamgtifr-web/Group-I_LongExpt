```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks
```

```python
data = pd.read_csv(
    "GROUPI_LAB0001.trt",
    sep=";",
    skiprows=300,
    skipfooter=300,
    names=["Wavelength_nm", "Counts"],
    engine="python"
)

x = data["Wavelength_nm"].values
y = data["Counts"].values

peaks, properties = find_peaks(
    y,
    height=2000, 
    distance=5         
)


plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Spectrum")
plt.plot(x[peaks], y[peaks], "ro", label="Peaks")

for p in peaks:
    plt.text(
        x[p], y[p]+200,
        f"{x[p]:.1f}",
        fontsize=10,        
        fontweight='bold',  
        ha='center',
        va='bottom'
    )
    
plt.margins(y=0.15)
plt.xlabel("Wavelength (nm)")
plt.ylabel("Counts")
plt.title("M-Normal Source On")
plt.legend()
plt.tight_layout()
plt.savefig("M-NormalSource_On_with_peaks.jpg", dpi=400)
plt.show()
```

```python

trtfile = pd.read_csv("GROUPI_LAB0003.trt")
display(trtfile)

data = pd.read_csv(
    "GROUPI_LAB0003.trt",
    sep=";",
    skiprows=8,
    names=["Wavelength_nm", "Counts"]
)

x = data["Wavelength_nm"].values
y = data["Counts"].values


minima, properties = find_peaks(
    -y,
    distance=10,
#    height=30
    prominence=250
)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label="Spectrum")
plt.plot(x[minima], y[minima], "ro", label="Minima")


for p in minima:
    plt.text(
        x[p], y[p] - 200,
        f"{x[p]:.1f}",
        fontsize=12,
        fontweight="bold",
        ha="center",
        va="top"
    )

plt.xlabel("Wavelength (nm)")
plt.ylabel("Counts")
plt.title("M-Sun (Absorption Spectrum)")
plt.legend()
plt.margins(y=0.15)
plt.tight_layout()
plt.savefig("M-Sun_absorption_with_minima.jpg", dpi=400)
plt.show()
```
